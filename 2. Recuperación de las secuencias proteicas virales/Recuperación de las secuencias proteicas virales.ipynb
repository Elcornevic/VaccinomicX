# Recuperación de las secuencias proteicas virales
Las secuencias proteicas de las <u>10 poliproteínas virales del DENV-2</u> se recuperaron de la base de datos del '**Universal Protein Resource**' - *Uniprot* (https://www.uniprot.org/) (*Uniprot Consortium, 2023*), a partir de una poliproteína de *3 391 aminoácidos* del genoma de DENV-2 (código de acceso: __Q9WDA6__) procedente de Iquitos y colectada en 1996, la cual es la única poliproteína en Perú que ha sido revisada en Uniprot hasta el momento, formando parte de la sección '*UniProtKB/Swiss-Prot*'.

```python
# Importar paquete de librería estándar (io), el módulo de 'bioservices' y 'Pandas'
import io
from bioservices import UniProt
import pandas as pd
```

```python
# Conectar a UniProt
service = UniProt()
```

```python
# Construir y realizar la consulta
query = 'Q9WDA6'
result = service.search(query)
print(result)
```

```python
##  Convertir el último resultado de búsqueda en un marco de datos en Pandas
df_table = pd.read_table(io.StringIO(result))
df_table
```

```python
# Visualizar la secuencia aminoacídica completa del genoma de la poliproteína del DENV-2
result_fasta = service.search(query, frmt = 'fasta')
print(result_fasta)
```

```python
## Guardar la secuencia aminoacídica completa del genoma de la poliproteína del DENV-2 (Formato fasta)
result_fasta = service.search(query, frmt = 'fasta')
file_genoma = 'Q9WDA6 - DENV_2 - genome polyprotein'

with open(file_genoma, 'w') as file:
    # Write the content to the file
    file.write(result_fasta)
```

```python
# Recuperar, visualizar y guardar cada una de las secuencias aminoacídicas de las proteínas del DENV-2 (Formato fasta)
result_fasta = service.search(query, frmt = 'fasta')

file_secuencias = 'Uniprot_Proteínas virales_DENV-2'

with open(file_secuencias, 'w') as file:
    print('>ancC_Proteína de la cápside anclada', file = file)
    print(result_fasta[111:212], file = file)
    print(file = file)
    print('>prM_Proteína premembrana', file = file)
    print(result_fasta[226:395], file = file)
    print(file = file)
    print('>E_Proteína de envoltura', file = file)
    print(result_fasta[395:898], file = file)
    print(file = file)
    print('>NS1_Proteína no estructural 1', file = file)
    print(result_fasta[898:1256], file = file)
    print(file = file)
    print('>NS2A_Proteína no estructural 2A', file = file)
    print(result_fasta[1256:1478], file = file)
    print(file = file)
    print('>NS2B_Proteína no estructural 2B', file = file)
    print(result_fasta[1478:1610],file = file)
    print(file = file)
    print('>NS3_Proteína no estructural 3', file = file)
    print(result_fasta[1610:2238], file = file)
    print(file = file)
    print('>NS4A_Proteína no estructural 4A', file = file)
    print(result_fasta[2238:2368], file = file)
    print(file = file)
    print('>NS4B_Proteína no estructural 4B', file = file)
    print(result_fasta[2391:2643], file = file)
    print(file = file)
    print('>NS5_Proteína no estructural 5', file = file)
    print(result_fasta[2643:3558], file = file)
    
with open('Uniprot_Proteínas virales_DENV-2') as f:
    contents = f.read()
    print(contents)
```
