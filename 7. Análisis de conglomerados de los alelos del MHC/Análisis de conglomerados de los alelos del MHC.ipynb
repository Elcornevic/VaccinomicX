# Análisis de conglomerados de los alelos del MHC
La herramienta en línea **MHCcluster v2.0** (https://services.healthtech.dtu.dk/services/MHCcluster-2.0/) (*Thomsen et al., 2013*) se utilizó para la predicción y el análisis de conglomerados de los posibles alelos MHC de clase I y clase II que pueden interactuar con los epítopos seleccionados durante las respuestas inmunitarias.

```python
# Importar los módulos necesarios de 'selenium'
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import Select
import time
import urllib
import urllib.request
from PIL import Image
```

```python
# Establecer la ubicación del WebDriver
s = Service('/usr/local/bin/chromedriver')
```

```python
# Inicializar el Selenium WebDriver
driver = webdriver.Chrome(service = s)

# Visitar la herramienta en línea MHCcluster v2.0
driver.get('https://services.healthtech.dtu.dk/services/MHCcluster-2.0/')
time.sleep(1)


# Realizar el análisis de conglomerados de los alelos del MHC clase I
driver.find_element(by = By.XPATH, value = '//*[@id="cookiescript_accept"]').click()
time.sleep(4)
alelos_mhc_i = driver.find_element(by = By.XPATH, value = '/html/body/div[3]/main/div/div[3]/div/div[2]/div/form/div/b/input[3]')
alelos_mhc_i.send_keys('/home/victor/Escritorio/Tesis/RESULTADOS/7. Análisis de conglomerados de los alelos del MHC/MHC_I')
## Entregar la data
send_analysis = driver.find_element(By.XPATH, '/html/body/div[3]/main/div/div[3]/div/div[2]/div/form/b/p[2]/input[1]')
driver.execute_script('arguments[0].click();', send_analysis)
time.sleep(90)

## Obtener el análisis
get_url = driver.current_url

img_tree = driver.find_element(by = By.XPATH, value = '/html/body/table/tbody/tr/td[1]/img')
src_t = img_tree.get_attribute('src')
urllib.request.urlretrieve(src_t, 'tree_mhci.png')


img_heat_map = driver.find_element(by = By.XPATH, value = '/html/body/table/tbody/tr/td[2]/img')
src_h = img_heat_map.get_attribute('src')
urllib.request.urlretrieve(src_h, 'heat_map_mhci.png')

print('MHC_I images downloaded!')
    
driver.close()
```
